```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(school_data_df, student_data_df, how="left", on=["school_name", "school_name"])
school_data_complete_df.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Count Schools
school_grouped_df = school_data_complete_df.groupby(['school_name'])
total_schools = (school_grouped_df["school_name"].nunique()).sum()
# total_schools
```

```python
#Count students
total_students = school_data_complete_df['Student ID'].count()
# total_students
```

```python
#Total Budget
total_budget = (school_grouped_df["budget"].unique()).sum()
# total_budget
```

```python
#Calculate the average math score
avg_math_score = school_data_complete_df['math_score'].mean()
# avg_math_score
```

```python
#Calculate the average reading scor
avg_reading_score = school_data_complete_df['reading_score'].mean() 
# avg_reading_score
```

```python
#Calculate the percentage of students with a passing MATH score (70 or greater)
math_s = school_data_complete_df['math_score'] >= 70
math_pass = (math_s.sum())
perc_math_pass = (math_pass/total_students)*100
# perc_math_pass
```

```python
#Calculate the percentage of students with a passing READING score (70 or greater)
reading_s = school_data_complete_df['reading_score'] >= 70
reading_pass = (reading_s.sum())
perc_reading_pass = (reading_pass/total_students)*100
# perc_reading_pass
```

```python
#Calculate the percentage of students who passed math and reading (% Overall Passing)
overall = (school_data_complete_df['math_score'] >= 70) &  (school_data_complete_df['reading_score'] >= 70)
overall_pass = (overall.sum())
perc_overall_pass = (overall_pass/total_students)*100
# perc_overall_pass
```

```python
# District Summary

#Creating summary table

District_Summary_df = pd.DataFrame({"Total Schools":[total_schools], 
                                    "Total Students":[total_students],
                                    "Total Budget":[total_budget],
                                    "Avg Math Score":[avg_math_score],
                                    "Avg Reading Score":[avg_reading_score],
                                    "% Passing Math":[perc_math_pass],
                                    "% Passing Reading":[perc_reading_pass],
                                    "% Overall Passing":[perc_overall_pass]})


#Formating summary table

District_Summary_df['Total Students'] = District_Summary_df['Total Students'].astype(int).map("{:,.0f}".format)
District_Summary_df['Total Budget'] = District_Summary_df['Total Budget'].astype(float).map("${:,.2f}".format)
District_Summary_df['Avg Math Score'] = District_Summary_df['Avg Math Score'].astype(float).map("{:.3f}".format)
District_Summary_df['Avg Reading Score'] = District_Summary_df['Avg Reading Score'].astype(float).map("{:.3f}".format)
District_Summary_df['% Passing Math'] = District_Summary_df['% Passing Math'].astype(float).map("{:.3f}".format)
District_Summary_df['% Passing Reading'] = District_Summary_df['% Passing Reading'].astype(float).map("{:.3f}".format)
District_Summary_df['% Overall Passing'] = District_Summary_df['% Overall Passing'].astype(float).map("{:.3f}".format)


District_Summary_df
```

```python
#Calculate the percentage of students with a passing MATH score (70 or greater) - PER SCHOOL
math_df = school_data_complete_df.query('math_score>=70')
math_s_group = math_df.groupby(['school_name'])
perc_math_pass_per_school = (math_s_group['math_score'].count()/total_students_per_school)*100
# perc_math_pass_per_school
```

```python
#Calculate the percentage of students with a passing READING score (70 or greater)- PER SCHOOL
reading_df = school_data_complete_df.query('reading_score>=70')
reading_s_group = reading_df.groupby(['school_name'])
perc_reading_pass_per_school = (reading_s_group['reading_score'].count()/total_students_per_school)*100
# perc_reading_pass_per_school
```

```python
#Calculate the percentage of students who passed math and reading (% Overall Passing)- PER SCHOOL
overall_df = school_data_complete_df.query('reading_score >=70 & math_score >= 70')
overall_s_group = overall_df.groupby(['school_name'])
perc_overall_pass_per_school = (overall_s_group['math_score'].count() /total_students_per_school)*100
# perc_overall_pass_per_school
```
